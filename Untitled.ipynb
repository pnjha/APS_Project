```python
import numpy as np
import pygraphviz as pgv
import networkx as nx
import matplotlib.pyplot as plt
import itertools

G = nx.Graph()
G.add_nodes_from([1,5])

x,y = np.loadtxt('planar.txt', delimiter = " ", unpack = True)

for x1,y1 in itertools.izip(x,y):
    G.add_edge(x1,y1)
    
nx.draw(G, with_labels=True)    

# plt.plot(x,y)
# plt.show()
```
